```python
sql_query_q3_5a = '''
SELECT 
s1.major, 
avg(s1.gpa) 
FROM students s1 
GROUP BY 
s1.major;'''
```

6A

```python
sql_query_q3_6a = '''select * from 
(
  SELECT * FROM (SELECT s1.* FROM students s1 WHERE s1.major = "Computer Science" ORDER BY s1.gpa DESC LIMIT 2)  
  UNION ALL 
  SELECT * FROM (SELECT * FROM students s2 WHERE s2.major = "Physics" ORDER BY s2.gpa DESC LIMIT 2)  
  UNION ALL 
  SELECT * FROM (SELECT * FROM students s3 WHERE s3.major = "Engineering" ORDER BY s3.gpa DESC LIMIT 2)
  ) o ORDER BY 
  o.major ASC, 
  o.gpa DESC ;'''
```

```python
pd.read_sql_query(sql_query_q3_6a,con = connection)
```
